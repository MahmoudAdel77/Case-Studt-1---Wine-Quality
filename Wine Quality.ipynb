```python
bin_edges = [2.720000, 3.110000, 3.210000, 3.320000, 4.010000]
bin_names = ['High', 'Moderately High', 'Medium', 'Low']
```

```python
wine_df['acidity_levels'] = pd.cut(wine_df['pH'], bin_edges, labels=bin_names)
wine_df.tail(20)
```

```python
wine_df.groupby(['acidity_levels'])['quality'].mean()
```

```python
wine_df.to_csv('winequality_edited.csv', index=False)
```

```python
low_alcohol = wine_df.query('alcohol < 10.3')
high_alcohol = wine_df.query('alcohol >= 10.3')
```

```python
mean_quality_low = low_alcohol.quality.mean()
```

```python
mean_quality_high = high_alcohol.quality.mean()
```

```python
low_sugar = wine_df[wine_df['residual sugar'] < 3.0]
high_sugar = wine_df[wine_df['residual sugar'] >= 3.0]
```

```python
num_samples == low_sugar['quality'].count() + high_sugar['quality'].count()
```

```python
mean_quality_low = low_sugar.quality.mean()
mean_quality_high = high_sugar.quality.mean()
```

```python
locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating')
```

```python
color_counts = wine_df.groupby(['color', 'quality']).count()['pH']
color_counts
```

```python
color_totals = wine_df.groupby('color').count()['pH']
color_totals
```

```python
red_proportions = color_counts['red'] / color_totals['red']
red_proportions
```

```python
white_proportions = color_counts['white'] / color_totals['white']
white_proportions
```

```python
red_proportions['9'] = 0
red_proportions
```

```python
ind = np.arange(len(red_proportions))
width = 0.35
```

```python
red_bars = plt.bar(ind, red_proportions, width, color='r', alpha=.7, label= 'Red Wine')
white_bars = plt.bar(ind + width, white_proportions, width, color='w', alpha=.7, label='White Wine')
plt.xlabel('Proportions')
plt.ylabel('Qulaity')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width / 2
labels = ['3', '4', '5', '6', '7', '8', '9']
plt.xticks(locations, labels)
plt.legend();
```
